```python
import nltk
from konlpy.corpus import kobill
import pandas as pd
files_ko = kobill.fileids()
```

```python
locations = ['cheonan', 'dongtan', 'guro', 'gwanghwamun', 'haeundae', 
             'jeju', 'mapo', 'samsung', 'seocho', 'seodaemun', 'ulsan', 'yeoksam']

data = pd.DataFrame()

for loc in locations:
    file = "./data/hotel_review_shillastay_" + loc + "_final.csv"
    tmp = pd.read_csv(file)
    tmp = tmp[['nickname', 'score', 'review', 'stay_date']]
    tmp['location'] = loc
    data = pd.concat([data, tmp], axis=0, ignore_index=True)
```

```python
data
```

```python
data1 = data.drop(['location'], axis=1)
data1
```

```python
data1.to_csv("./data/shillastay.csv", encoding="utf-8-sig")
```

```python
data_review = list(data.review)
```

```python
data_one = ' '.join(data_review)
data_one
```

```python
from konlpy.tag import Okt
t = Okt()
tokens_re = t.nouns(data_one)
tokens_re
```

```python
rev = nltk.Text(tokens_re, name='신라스테이')
print(len(set(rev.tokens)))
rev.vocab()
```
